```python
import numpy as np
import pure_cv as vc
import cv2 as cv
```

```python
bboxes = np.array([
  [[100, 100], [400, 150]],
  [[200, 200], [300, 300]],
])
bg = vc.black(512, 512)
vc.show(
  vc.draw.bboxes(
    vc.draw.bboxes(bg, bboxes),
    vc.pad_bboxes(bboxes, l=0.1, r=0.1, t=0.1, b=0.1), color=(0, 255, 0)
  )
)
```
